```python
data = [[]]
with open('particle_pos.txt', encoding='utf8') as f:
    for line in f:
        a = line.strip().split()
        b = [float(x) for x in a]
        data.append(b)
```

```python
import matplotlib.pyplot as plt
  
def plot_particles(x):
    y = [0 for i in range(len(x))]
    plt.scatter(x, y)
    plt.show()

    
for d in data:
    plot_particles(d)
```
